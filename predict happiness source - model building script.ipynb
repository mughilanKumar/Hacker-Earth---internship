```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
print(os.listdir())
```

```python
dataframe = pd.read_csv('hm_train.csv')
dataframe.head()
```

```python
dataframe.info()
```

```python
df = dataframe.loc[:,('cleaned_hm','predicted_category')]
```

```python
df.head()
```

```python
x = df.loc[:,'cleaned_hm']
y = df.loc[:,'predicted_category']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words='english')
```

```python
x_cv = cv.fit_transform(x)
```

```python
x_cv.shape
```

```python
x_names = cv.get_feature_names()
x_names
```

```python
X = pd.DataFrame(x_cv.toarray(), columns=x_names)
```

```python
X.shape
```

```python
Y = y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc = RandomForestClassifier(criterion='gini',random_state=0)
rc.fit(x_train,y_train)
y_pred = rc.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.accuracy_score(y_test,y_pred))
```
